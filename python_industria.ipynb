Em outras palavras, é um registro da quantidade de dinheiro que cada setor econômico (identificado pela sua CNAE) arrecadou em um determinado mês, discriminando-se os tributos que foram pagos através do Darf.

#Proposta              
é possível analisar a arrecadação bruta mensal discriminada por tributo e setor econômico, identificado pela seção da CNAE, além da evolução da arrecadação ao longo do tempo para cada setor econômico.

**Arrecadação de IR e IPI detalhada**                
Contém os dados de arrecadação decendial do Imposto sobre a Renda (IR) e do Imposto sobre Produtos Industrializados (IPI), discriminados por arrecadação bruta, os respectivos eventos que afetam essa arrecadação bruta e a arrecadação líquida, valor este último passível de repasse aos Fundos de Participação dos Estados e do Municípios, FPM e FPE, de acordo com dispositivo constitucional.

#Resumo                 
"Arrecadação de IR e IPI detalhada" é um conjunto de dados que apresenta a arrecadação decendial (ou seja, a cada dez dias) do Imposto sobre a Renda (IR) e do Imposto sobre Produtos Industrializados (IPI), discriminados por arrecadação bruta, os respectivos eventos que afetam essa arrecadação bruta e a arrecadação líquida. Essa arrecadação líquida é o valor que pode ser repassado aos Fundos de Participação dos Estados e do Municípios (FPM e FPE), de acordo com dispositivo constitucional.


#Proposta                
Esses dados permitem analisar a arrecadação bruta e líquida de IR e IPI a cada dez dias, assim como identificar os eventos que afetaram a arrecadação bruta nesse período. Além disso, também é possível avaliar o impacto da arrecadação líquida no repasse aos fundos estaduais e municipais, contribuindo para o monitoramento do orçamento público e para a tomada de decisões em políticas fiscais.

```python
# transformando pano e mes para tipo data

meses_numerico = {
    'Janeiro':'01',
    'Fevereiro':'02',
    'Março':'03',
    'Abril':'04',
    'Maio':'05',
    'Junho':'06',
    'Julho':'07',
    'Agosto':'08',
    'Setembro':'09',
    'Outubro':'10',
    'Novembro':'11',
    'Dezembro':'12',
}

# Criar uma nova coluna 'data' combinando as informações das colunas 'ano' e 'mes'
ir_ipi['data'] = ir_ipi.apply(lambda row: pd.to_datetime(str(row['Ano']) + '-' + meses_numerico[row['Mês']] + '-01'), axis=1)
```

```python
ir_ipi = ir_ipi.applymap(lambda ir_ipi: ir_ipi.replace(',', '') if isinstance(ir_ipi, str) else ir_ipi)
ir_ipi_for = ir_ipi.drop(['data','Tributo','Ano','Mês','Decêndio'],axis =1).astype(float)
ir_ipi_for['Data'] = ir_ipi['Ano']
ir_ipi_for['Tributo'] = ir_ipi['Tributo']
# tabela.drop(['Coluna'], axis = 1, inplace=True)
```

```python
ir_ipi_for.info()
```

```python
# Agrupar por ano e calcular a soma dos valores de cada imposto
resultado = ir_ipi_for[['Data','Tributo','Arrecadação Líquida']].groupby(['Data','Tributo']).sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# filtrar os dados por ano e criar um gráfico separado para cada ano
anos_unicos = resultado.index.get_level_values('Data').unique()


for ano in anos_unicos:
    dados_ano = resultado.loc[ano]
    fig, ax = plt.subplots(figsize=(8, 7))
    
    # Usar o estilo seaborn no gráfico de barras
    sns.barplot(data=dados_ano, x=dados_ano.index, y="Arrecadação Líquida", ax=ax)
    
    plt.title(f'Arrecadação Líquida por Tributo - Ano {ano}')
    plt.xlabel('Data, Tributo')
    plt.ylabel('Arrecadação Líquida')     
    # Adicionar os valores acima das colunas
    for i, valor in enumerate(dados_ano["Arrecadação Líquida"]):
         ax.annotate(f'{valor:.2f}', (i, valor), xytext=(0, 5), textcoords='offset points', ha='center', va='bottom')
    plt.show()
```
