**Arrecadação de IR e IPI detalhada**                
Contém os dados de arrecadação decendial do Imposto sobre a Renda (IR) e do Imposto sobre Produtos Industrializados (IPI), discriminados por arrecadação bruta, os respectivos eventos que afetam essa arrecadação bruta e a arrecadação líquida, valor este último passível de repasse aos Fundos de Participação dos Estados e do Municípios, FPM e FPE, de acordo com dispositivo constitucional.

#Resumo                 
"Arrecadação de IR e IPI detalhada" é um conjunto de dados que apresenta a arrecadação decendial (ou seja, a cada dez dias) do Imposto sobre a Renda (IR) e do Imposto sobre Produtos Industrializados (IPI), discriminados por arrecadação bruta, os respectivos eventos que afetam essa arrecadação bruta e a arrecadação líquida. Essa arrecadação líquida é o valor que pode ser repassado aos Fundos de Participação dos Estados e do Municípios (FPM e FPE), de acordo com dispositivo constitucional.


#Proposta                
Esses dados permitem analisar a arrecadação bruta e líquida de IR e IPI a cada dez dias, assim como identificar os eventos que afetaram a arrecadação bruta nesse período. Além disso, também é possível avaliar o impacto da arrecadação líquida no repasse aos fundos estaduais e municipais, contribuindo para o monitoramento do orçamento público e para a tomada de decisões em políticas fiscais.

```python
#transformando algumas colunas tipo float

filtrando = tabela_Arrecadação_Estado.drop(['Ano','Mês'],axis=1)
#colunas numéricas para conversão
```

```python
#recadação de cada estado de cada mês
#filtrar os dados 

#substituindo valores vazio por zero
filtrando.fillna(0, inplace=True)
```

```python
#transfomando toda coluna em tipo float para depois somar os valores

# substituir todas as vírgulas em todas as colunas
filtrando = filtrando.applymap(lambda x: x.replace(',', '') if isinstance(x, str) else x)
filtrando = filtrando.applymap(lambda x: x.replace('.', '') if isinstance(x, str) else x)

# configuração de formatação do pandas para números de ponto flutuante
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
#transformando todas as colunas de string para float
# filtrando.apply(pd.to_numeric, errors='coerce')
# filtrando.info()

#iterar sobre as colunas do DataFrame e converter para tipo float
for col in filtrando.columns:
    filtrando[col] = pd.to_numeric(filtrando[col], errors='coerce')
filtrando.info()
```

```python
#criando uma coluna soma e somando os imposto de cada linha.
filtrando["Total"] = filtrando.sum(axis = 1)

#agora agrupar por estado

# mais antes criar o grupo por estado, adionar a coluna estado
filtrando["UF"] = tabela_Arrecadação_Estado["UF"]
```

```python
#criando grupo por estado 
indice_estado = filtrando[["UF","Total"]].groupby("UF").sum()

# agora vamos criar um grafico mais como os númeors estão grandes vou criar outra coluna mostrando o porcentual de cada um e depois aplicar no grafico
indice_estado["Porcentagem"] = (indice_estado["Total"]/(indice_estado["Total"].sum()))*100

display(indice_estado)
display(indice_estado["Porcentagem"].sum())
```

```python
indice_estado.drop(['Total'], axis = 1, inplace=True)
```

```python
#grafico
import matplotlib.pyplot as plt

#criando grafico de linha
fig, ax = plt.subplots(figsize=(12, 8))

x = indice_estado.index
y = indice_estado["Porcentagem"]
#passandos os dados 
ax.plot(x,y)

#personalizando o grafico adicionando rotulos para eixo x e para eixo y e titulo
ax.set_xlabel("Estados")
ax.set_ylabel("Porcentagem")
ax.set_title("Mostrando o valor de cada arrecadação por estado")


#Adicionar os valores em cima das barras
# for i, valor in enumerate(indice_estado["Porcentagem"]):
#     ax.annotate(f'{valor:.2f}', (i, valor), ha='center', va='bottom')
    
plt.show()
```
