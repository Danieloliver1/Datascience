```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_consumo = pd.read_csv(r"https://raw.githubusercontent.com/Danieloliver1/Datascience/main/Consumption_Total.csv")
x_producao = pd.read_csv(r"https://raw.githubusercontent.com/Danieloliver1/Datascience/main/Production_Total.csv")
```

```python
x_producao.head()
```

```python
# criando uma tabela 
total_consumo = x_consumo[["2010", "2011","2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019","2020", "2021","Total_Consumption"]]
total_producao = x_producao[["2010", "2011","2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019","2020", "2021","Total_production"]]
```

```python
# removendo caracteres comsumo
total_consumo = total_consumo.applymap(lambda x: x.replace('-','0') if isinstance(x, str) else x)
# removendo caracteres producao
total_producao = total_producao.applymap(lambda x: x.replace('-','0') if isinstance(x, str) else x)
```

```python
# Substitua os valores vazios comsumo com 0
total_consumo = total_consumo.fillna(0)
# Substitua os valores vazios producao com 0
total_producao = total_producao.fillna(0)
```

```python
# transformado o tipo numerico cosumo
total_consumo = total_consumo.astype(float)
# transformado o tipo numerico cosumo
total_producao = total_producao.astype(float)
```

```python
# enviando coluna continente para cosumo
total_consumo['Continent'] = x_consumo['Continent']
# enviando coluna continente para producao
total_producao['Continent'] = x_producao['Continent']
```

```python
# criando uma coluna porcentagem total de consumo de cada continente 
total_consumo["Porcentagem"] = (total_consumo["Total_Consumption"]/(total_consumo["Total_Consumption"].sum())) * 100
total_consumo['Porcentagem'] = total_consumo['Porcentagem'].apply(lambda x: round(x, 2))

# criando uma coluna porcentagem total de producao de cada continente 
total_producao["Porcentagem"] = (total_producao["Total_production"]/(total_producao["Total_production"].sum())) * 100
total_producao['Porcentagem'] = total_producao['Porcentagem'].apply(lambda x: round(x, 2))
```

```python

x_consumo = total_consumo.groupby('Continent')['Porcentagem'].agg('sum')
x_producao = total_producao.groupby('Continent')['Porcentagem'].agg('sum')

```

```python
#criando grafico de radar para comsumo
# Calcular os ângulos para os eixos do gráfico
num_categorias = len(x_consumo)
angulos = np.linspace(0, 2*np.pi, num_categorias, endpoint=False).tolist()
angulos += angulos[:1]  # Adicionar o primeiro ângulo no final para fechar o polígono
# Preparar os valores para o plot
valores = x_consumo.tolist()
valores += valores[:1]  # Adicionar o primeiro valor no final para fechar o polígono
# Configurar a figura e os eixos
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
# Plotar o gráfico de radar
ax.plot(angulos, valores, color='blue', linewidth=1)
ax.fill(angulos, valores, color='blue', alpha=0.25)
# Configurar os rótulos das categorias
ax.set_xticks(angulos[:-1])
ax.set_xticklabels(x_consumo.index)
ax.yaxis.grid(True)
# Adicionar um título
ax.set_title('Total de consumo por continente', size=14, weight='bold')
# Exibir o gráfico
plt.show()
```

```python
#criando grafico de radar para producao
# Calcular os ângulos para os eixos do gráfico
num_categorias = len(x_producao)
angulos = np.linspace(0, 2*np.pi, num_categorias, endpoint=False).tolist()
angulos += angulos[:1]  # Adicionar o primeiro ângulo no final para fechar o polígono
# Preparar os valores para o plot
valores = x_producao.tolist()
valores += valores[:1]  # Adicionar o primeiro valor no final para fechar o polígono
# Configurar a figura e os eixos
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
# Plotar o gráfico de radar
ax.plot(angulos, valores, color='blue', linewidth=1)
ax.fill(angulos, valores, color='blue', alpha=0.25)
# Configurar os rótulos das categorias
ax.set_xticks(angulos[:-1])
ax.set_xticklabels(x_producao.index)
ax.yaxis.grid(True)
# Adicionar um título
ax.set_title('Total de consumo por continente', size=14, weight='bold')
# Exibir o gráfico
plt.show()
```

```python
#grafico
import matplotlib.pyplot as plt


# Criar uma figura com tamanho específico
fig = plt.figure(figsize=(13, 7))
# x = x_consumo.index
# y = x_consumo
#passandos os dados 
plt.plot(x_consumo.index,x_consumo,label='Consumo')
plt.plot(x_consumo.index,x_producao,label='Produção')

# Configurar o título e os rótulos dos eixos
plt.title('Consumo e Produção ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Quantidade')

# Definir limites dos eixos x e y
plt.xlim(x_consumo.index[0], x_consumo.index[-1])
plt.ylim(min(x_consumo.min(), x_producao.min()), max(x_consumo.max(), x_producao.max()))

# Adicionar legenda
plt.legend()
    
plt.show()
```

```python
# excluindo coluna porcentagem comsumo
total_consumo.drop(['Porcentagem'], axis = 1, inplace=True)
total_consumo.drop(['Total_Consumption'], axis = 1, inplace=True)

# excluindo coluna porcentagem producao
total_producao.drop(['Porcentagem'], axis = 1, inplace=True)
total_producao.drop(['Total_production'], axis = 1, inplace=True)
```

```python
# Usar a função melt() para converter as colunas de anos em uma coluna

df_consumo = pd.melt(total_consumo, id_vars='Continent', var_name='Ano', value_name='Valor')
df_producao = pd.melt(total_producao, id_vars='Continent', var_name='Ano', value_name='Valor')
```

```python
agrupando_consumo = df_consumo.groupby("Ano")
agrupando_producao = df_producao.groupby("Ano")
# agrupando_consumo.head()
```

```python
 #quantidade de produção por ano 
for ano, grupo in agrupando_consumo:
    contagem = grupo['Continent'].value_counts()
    print("Ano:", ano)
    print(contagem)
```

```python
# Contar a quantidade de produção de cada ano
for ano, grupo in agrupando_producao:
    contagem = grupo['Continent'].value_counts()
    fig = plt.figure(figsize=(13, 7))
    plt.bar(contagem.index, contagem.values)
    plt.title(f"Contagem de produção de cada ano em {ano}")
    plt.xlabel("Ano")
    plt.ylabel("Quantidade")
    plt.show()
```

```python
 #quantidade de consumo por ano
for ano, grupo in agrupando_producao:
    contagem = grupo['Continent'].value_counts()
    print("Ano:", ano)
    print(contagem)
```

```python
# Contar a quantidade de consumo de cada ano
for ano, grupo in agrupando_consumo:
    contagem = grupo['Continent'].value_counts()
    fig = plt.figure(figsize=(13, 7))
    plt.bar(contagem.index, contagem.values)
    plt.title(f"Contagem de consumo de cada ano em {ano}")
    plt.xlabel("Ano")
    plt.ylabel("Quantidade")
    plt.show()
```
